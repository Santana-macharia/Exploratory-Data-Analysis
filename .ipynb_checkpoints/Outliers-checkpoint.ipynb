```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
pd.options.display.max_rows = 10
pd.options.display.float_format = '{:,.2f}'.format #supress scientific notation
plt.rcParams['figure.figsize'] = (12, 8)
```

### Read Data

```python
loan = pd.read_csv("data/train.csv")
loan.head()
```

```python
loan['Loan_Status'] = np.where(loan['Loan_Status'] == 'Y' , 0, 1)
```

```python
print('Highest Applicant Income is:',loan['ApplicantIncome'].max())
print('Lowest Applicant Income is:',loan['ApplicantIncome'].min())
print('Average Applicant Income is:',loan['ApplicantIncome'].mean())
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
sns.violinplot(y="ApplicantIncome", data=loan,split=True,ax=ax[0])
ax[0].set_title('Violin Plot of Applicant Income')
sns.boxplot(y=loan['ApplicantIncome'],ax=ax[1])
ax[1].set_title('Boxplot of Applicant Income')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
loan[loan['Loan_Status']==0].ApplicantIncome.plot.hist(ax=ax[0],bins=20,edgecolor='black',color='red')
ax[0].set_title('Loan_Status= 0')
loan[loan['Loan_Status']==1].ApplicantIncome.plot.hist(ax=ax[1],color='green', bins=20,edgecolor='black')
ax[1].set_title('Loan_Status= 1')
plt.show()
```

```python
loan["ApplicantIncome"].fillna(loan["ApplicantIncome"].mean(), inplace=True)
```

[We will use this to get our outlierw](https://www.itl.nist.gov/div898/handbook/prc/section1/prc16.htm)

```python
#what is an outlier

# Calculate first and third quartile
first_quartile = loan['ApplicantIncome'].describe()['25%']
third_quartile = loan['ApplicantIncome'].describe()['75%']

#Interquartile range
iqr = third_quartile - first_quartile

#Remove outliers
loan_appinc_without_outliers = loan[(loan['ApplicantIncome'] > (first_quartile - 3 * iqr)) &
            (loan['ApplicantIncome'] < (third_quartile + 3 * iqr))]
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
sns.violinplot(y="ApplicantIncome", data=loan_appinc_without_outliers,split=True,ax=ax[0])
ax[0].set_title('Violin Plot of Applicant Income')
sns.boxplot(y=loan_appinc_without_outliers['ApplicantIncome'],ax=ax[1])
ax[1].set_title('Boxplot of Applicant Income')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
loan_appinc_without_outliers[loan_appinc_without_outliers['Loan_Status']==0].ApplicantIncome.plot.hist(ax=ax[0],
                                                                                           bins=20,edgecolor='black',color='red')
ax[0].set_title('Loan_Status= 0')
loan_appinc_without_outliers[loan_appinc_without_outliers['Loan_Status']==1].ApplicantIncome.plot.hist(ax=ax[1],
                                                                                           color='green',
                                                                                           bins=20,edgecolor='black')
ax[1].set_title('Loan_Status= 1')
plt.show()
```
